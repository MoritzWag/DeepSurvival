```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import os 
import seaborn as sns
```

```python
# retrieve tabular weights
path_lw = "../linear_weights"

features_list =  ['ABETA', 'APOE4', 'AV45-PET',
                  'Male',
                  'FDG-PET', 'p-Tau', 
                  'T-Tau', 'Age', 'Age:Male',
                  'Age_1', 'Age_2', 'Age_3', 'Age_4',
                  'Education (Linear)', 'Education (Quadratic)', 'Education (Cubic)']
weights = []

# tabular only
weights_tab_0 = np.load(file=f"{path_lw}/weights_0.npy")[ :len(features_list)]
weights.append(weights_tab_0)
weights_tab_1 = np.load(file=f"{path_lw}/weights_1.npy")[ :len(features_list)]
weights.append(weights_tab_1)
weights_tab_2 = np.load(file=f"{path_lw}/weights_2.npy")[ :len(features_list)]
weights.append(weights_tab_2)
weights_tab_3 = np.load(file=f"{path_lw}/weights_3.npy")[ :len(features_list)]
weights.append(weights_tab_3)
weights_tab_4 = np.load(file=f"{path_lw}/weights_4.npy")[ :len(features_list)]
weights.append(weights_tab_4)

# deep tabular weights wo orthogonalization
weights_deep_0 = np.load(file=f"{path_lw}/0/weights_deep_0.npy")[:, :len(features_list)].squeeze()
weights.append(weights_deep_0)
weights_deep_1 = np.load(file=f"{path_lw}/1/weights_deep_1.npy")[:, :len(features_list)].squeeze()
weights.append(weights_deep_1)
weights_deep_2 = np.load(file=f"{path_lw}/2/weights_deep_2.npy")[:, :len(features_list)].squeeze()
weights.append(weights_deep_2)
weights_deep_3 = np.load(file=f"{path_lw}/3/weights_deep_3.npy")[:, :len(features_list)].squeeze()
weights.append(weights_deep_3)
weights_deep_4 = np.load(file=f"{path_lw}/4/weights_deep_4.npy")[:, :len(features_list)].squeeze()
weights.append(weights_deep_4)

# deep tabular weights with orthogonalization
weights_deep_0_orth = np.load(file=f"{path_lw}/orth_0/weights_deep_0.npy")[:, :len(features_list)].squeeze()
weights.append(weights_deep_0_orth)
weights_deep_1_orth = np.load(file=f"{path_lw}/orth_1/weights_deep_1.npy")[:, :len(features_list)].squeeze()
weights.append(weights_deep_1_orth)
weights_deep_2_orth = np.load(file=f"{path_lw}/orth_2/weights_deep_2.npy")[:, :len(features_list)].squeeze()
weights.append(weights_deep_2_orth)
weights_deep_3_orth = np.load(file=f"{path_lw}/orth_3/weights_deep_3.npy")[:, :len(features_list)].squeeze()
weights.append(weights_deep_3_orth)
weights_deep_4_orth = np.load(file=f"{path_lw}/orth_4/weights_deep_4.npy")[:, :len(features_list)].squeeze()
weights.append(weights_deep_4_orth)



```

```python
weights = np.concatenate(weights)
weights
```

```python
model_type = np.array(["Linear", "Wide and Deep w/o orth", "Wide and Deep with orth"])
model_type = np.repeat(model_type, len(features_list)*5)
coefficients = np.array(features_list)
# coefficients = np.arange(len(features_list))
coefficients = np.tile(coefficients, 5*3)
model_type
```

```python
model_type.shape
```

```python
df_coeff = pd.DataFrame(columns=['model_type', 'feature', 'coefficient'])
df_coeff['model_type'] = model_type
df_coeff['feature'] = coefficients
df_coeff['coefficient'] = weights
#df_coeff = df_coeff.sort_values(by=["feature"])


df_coeff[210:]
```

```python
df_coeff['coefficient'].unique().shape
```

```python
plt.figure(figsize=(16, 10))
ax = sns.boxplot(y="feature", x="coefficient", hue="model_type", data=df_coeff, palette="husl", showfliers=False)
ax.get_legend().set_title(None)
plt.savefig("coefficients.png")
# ax = sns.boxplot(x="feature", y="coefficient", hue="model_type", data=df_coeff, showfliers=False, palette="husl")
```

```python
# retrieve c-index results
baseline = np.array([0.633, 0.593, 0.637, 0.6338, 0.65])
tabular = np.array([0.7315, 0.7476, 0.7476, 0.753, 0.7669])
wide_deep_orth = np.array([0.7661, 0.770, 0.750, 0.825, 0.803])
wide_deep = np.array([0.7715, 0.7502, 0.758, 0.762, 0.776])

# define dataframe 
df_results = pd.DataFrame(columns=["Linear", "Deep", "Wide and Deep w/o orth", "Wide and Deep w/ orth"])
df_results['Linear'] = tabular
df_results['Deep'] = baseline
df_results['Wide and Deep w/o orth'] = wide_deep
df_results['Wide and Deep w/ orth'] = wide_deep_orth
df_results.head()
# build boxplot
```

```python
df_results2 = pd.DataFrame(columns=["Model", "c-index"])
models = np.array(["Linear", "Deep", "Wide and Deep w/o orth", "Wide and Deep w/ orth"])
models = np.repeat(models, 5)
c_index = np.concatenate([tabular, baseline, wide_deep, wide_deep_orth])

df_results2["Model"] = models
df_results2['c-index'] = c_index

plt.figure(figsize=(8, 8))
ax = sns.boxplot(y="c-index", x="Model", data=df_results2, palette="husl", showfliers=False)
ax.set_xticklabels(["Linear", "Deep", "Wide and Deep \n w/o orth", "Wide and Deep \n w/ orth"])
plt.savefig("results.png")

#ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
```
